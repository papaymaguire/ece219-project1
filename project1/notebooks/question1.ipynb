```python
import matplotlib.pyplot as plt
dataset['full_text_char_count'] = dataset['full_text'].str.len()
dataset['full_text_char_count'].hist(bins=20)
plt.title("Occurrences of Document Full Text Length")
plt.xlabel("Full Text Char Count")
plt.ylabel("Occurrences")
dataset['full_text_char_count'].max()
```

```python
dataset['leaf_label'].value_counts().plot(kind="bar")
plt.title("Occurrences of Document Leaf Label Classification")
plt.xlabel("Leaf Label")
plt.ylabel("Occurrences")
```

```python
dataset['root_label'].value_counts().plot(kind="bar")
plt.title("Occurrences of Document Root Label Classification")
plt.xlabel("Root Label")
plt.ylabel("Occurrences")
```
